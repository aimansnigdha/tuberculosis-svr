```python
#convert 28x28 grayscale to 48x48 rgb channels
def to_rgb(img): 
    img_rgb = np.asarray(np.dstack((img, img, img)))
    img = cv2.resize(img_rgb, (299, 299), interpolation = cv2.INTER_AREA)
    return img
```
